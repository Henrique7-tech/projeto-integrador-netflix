```python

# Questionamento: Em quais anos a Netflix adicionou mais conteúdos?

# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando o dataset
# Delimitando caminhos
df = pd.read_csv(r"C:\Users\Andrey\Documents\Faculdade\ProjetoIntegradorDados\projeto-integrador-netflix\dados\processed\netflix_clean.csv")

# Verificando os dados
print(df.head())
```

```python
# Conteúdos adicionados por ano

# Conta quantos títulos foram adicionados em cada ano
conteudos_por_ano = (
    df['year_added']
    .value_counts()
    .sort_index()
)

print("\nConteúdos adicionados por ano:\n")
print(conteudos_por_ano)
```

```python
# Encontrando o ano com mais conteúdos


ano_max = conteudos_por_ano.idxmax()
qtd_max = conteudos_por_ano.max()

print(f"\nAno com mais conteúdos adicionados: {ano_max}")
print(f"Quantidade de conteúdos: {qtd_max}")
```

```python
# Visualização gráfica
plt.figure(figsize=(12,6))

plt.plot(
    conteudos_por_ano.index,
    conteudos_por_ano.values,
    marker='o'
)

plt.title("Conteúdos adicionados à Netflix por ano")
plt.xlabel("Ano")
plt.ylabel("Quantidade de conteúdos")

plt.xticks(conteudos_por_ano.index, rotation=45)

plt.grid(True)

plt.show()
```
