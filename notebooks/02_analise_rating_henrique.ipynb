```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando o arquivo que o Felipe limpou e salvou
df = pd.read_csv('../dados/processed/netflix_clean.csv')
```

```python
# Contando quantos títulos tem em cada classificação
contagem_rating = df['rating'].value_counts().reset_index()

# Renomeando as colunas para ficar organizado
contagem_rating.columns = ['Classificacao', 'Quantidade']

# O caractere '#' diz para o Python ignorar a linha e não printar o texto na tela
# contagem_rating
```

```python
# Configurando o estilo do grupo
sns.set_style("whitegrid")
plt.figure(figsize=(11, 6))

# Criando o gráfico de barras com Seaborn usando tons de vermelho da Netflix
sns.barplot(
    data=contagem_rating, 
    x='Classificacao', 
    y='Quantidade', 
    palette='Reds_r'
)

# Títulos e legendas do gráfico
plt.title('Perfil de Público do Catálogo Netflix (Rating)', fontsize=14, weight='bold', pad=15)
plt.xlabel('Classificação Indicativa', fontsize=12, labelpad=10)
plt.ylabel('Total de Títulos', fontsize=12, labelpad=10)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
