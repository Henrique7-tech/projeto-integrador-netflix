```python
import pandas as pd

# Carregando o arquivo que o Felipe limpou e salvou
df = pd.read_csv('../dados/processed/netflix_clean.csv')

# Exibe as primeiras 5 linhas para garantir que os dados abriram
df.head()
```

```python
# Contando quantos títulos tem em cada classificação
contagem_rating = df['rating'].value_counts().reset_index()

# Renomeando as colunas para ficar organizado
contagem_rating.columns = ['Classificacao', 'Quantidade']

# Mostra o resultado na tela
contagem_rating
```

```python
import plotly.express as px

# Criando o gráfico de barras com as cores da Netflix
fig = px.bar(contagem_rating, 
             x='Classificacao', 
             y='Quantidade',
             title='Perfil de Público do Catálogo Netflix (Rating)',
             labels={'Classificacao': 'Classificação Indicativa', 'Quantidade': 'Total de Títulos'},
             template='plotly_dark',
             color='Quantidade', 
             color_continuous_scale='Reds')

fig.show()
```
